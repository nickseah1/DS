# Function Below to graph varying vs constant scenarios

Parameters varying and constant are series

Age group is a string

Ylabel is a string that should represent what you are measuring (i.e. deaths, exposed, hospitalizations)

```python
def plot_func(varying,constant,agegroup, ylabel):

    fig = plt.figure(figsize = (8,3))
    
    Days = range(len(varying))


    plt.plot(Days, varying)
    plt.plot(Days, constant)
    plt.legend(['Varying Testing Rate', 'Constant Testing Rate'])
    plt.title('Varying vs Constant Testing Rates for ' + agegroup)

    plt.xlabel('Days')
    plt.ylabel(ylabel)


    # Remove
    plt.savefig('VaryingvsConstantScenario.png')
    return plt.show()
```

# Experiment

This experiment is to determine the effect of the infection rate on the trend we noted above. Specifically does swapping the senior and child infection rates still reduce the number of deaths and exposures when comparing varying and constant testing rate.

  Below we plot the child deaths and senior deaths from the STEM simulation model.

```python
child_swap_deaths = plot_func(Child_deaths_v, Child_deaths_c, 'Children', 'Deaths')
adult_swap_deaths = plot_func(Adult_deaths_v, Adult_deaths_c, 'Adults', 'Deaths')
senior_swap_deaths = plot_func(Senior_deaths_v, Senior_deaths_c, 'Seniors', 'Deaths')
```

# Analysis
Reading the graphs above we can tell that even when the child and senior infection rates are swapped, a varying testing rate where we allocate the majority of asymptomatic testing resouces to the Child and Adult Populations instead of the senior population results in a significant reduction in the total deaths for all age groups through 400 days of the pandemic.
