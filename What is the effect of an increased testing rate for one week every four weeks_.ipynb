# Does periodic COVID-19 testing result in lower infections and deaths compared to homogenous COVID-19 testing?

    This project is to determine if testing a massive group of asymptomatic individuals once every four weeks would result in fewer deaths and infections than testing the same amount of people at the same rate every week. The data comes from a Multi-population simulation model in the Spatial-Temporal Epidemiological Modeler (STEM). To ensure we are using the same number of tests for the two scenarios we set the average tests for the 4 weeks span to be equal.

```python
import pandas as pd
import matplotlib.pyplot as plt
```

The dataframes that begin with P stand for periodic, which means that the data came from a scenario where asymptomatic individuals are tested periodically. The dataframes that begin with NP stand for non-periodic, which means that the data came from a scenario where asymptomatic individuals were tested at a constant rate for the timeframe.

```python
def plot_func(Periodic,NonPeriodic,agegroup, ylabel):

    fig = plt.figure(figsize = (8,3))
    
    Days = range(len(Periodic))


    plt.plot(Days, Periodic)
    plt.plot(Days, NonPeriodic)
    plt.legend(['Periodic Testing', 'NonPeriodic Testing'])
    plt.title('Periodic vs NonPeriodic Testing for ' + agegroup)

    plt.xlabel('Days')
    plt.ylabel(ylabel)


    # Remove
    # plt.savefig('ExposedVaryingTestingRate.png')
    return plt.show()
```

# Experiment
  We are conducting an experiment in STEM that compares two scenarios. The first scenario has an asymptomatic testing rate of 0.07 for week 1 and an asymptomatic testing rate of 0.01 for weeks 2, 3, and 4. This starts on Day 100 and repeats itself once every four weeks until Day 156. The second scenario has non-periodic testing where phi_a is set to 0.02 from day 100 to 156.

```python
plot_func(P_ChildExposed, NP_ChildExposed, 'Children', 'Exposed')
plot_func(P_AdultExposed, NP_AdultExposed, 'Adults', 'Exposed')
plot_func(P_SeniorExposed, NP_SeniorExposed, 'Seniors', 'Exposed')
```

# Analysis
  From these graphs we can see that there isn't much of an effect in the number of people exposed when we increase the testing rate of asymptomatic individuals for one week the lower it for the next 3 weeks periodically
